```python
from utils import load_test_images, show_images
images = load_test_images(num_images=5, bw=True)
show_images(images)
```

# Step 1: Preprocess

```python
from utils import preprocess_image
images = [preprocess_image(img) for img in images]
show_images(images)
```

# Step 2: Canny Edge Detection

```python
from skimage import filter
edges = [filter.canny(img, sigma=3.5) for img in images] #sigma should be between like 3 and 5 for bold contours
show_images(edges)
```

# Step 3: Hough Transform

```python
from skimage import transform
img = edges[0]
hspace, angles, dists = transform.hough_line(img)
hspace, angles, dists = transform.hough_line_peaks(hspace, angles, dists)
# plt.imshow(img, cmap=plt.cm.gray)
# plt.title('Input image')

# plt.subplot(1, 2, 2)
# plt.imshow(out, cmap=plt.cm.bone,
#            extent=(d[0], d[-1],
#                    np.rad2deg(angles[0]), np.rad2deg(angles[-1])))
# plt.title('Hough transform')
# plt.xlabel('Angle (degree)')
# plt.ylabel('Distance (pixel)')

# plt.subplots_adjust(wspace=0.4)
# plt.show()
```
