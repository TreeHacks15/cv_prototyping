```python
import numpy
import scipy
import skimage
import matplotlib.pyplot as plt
%pylab inline
```

```python
from utils import load_test_images
images = load_test_images(num_images=5, bw=True)
```

# Step 1: Preprocess

```python
from utils import preprocess_image
images = [preprocess_image(img) for img in images]
plt.imshow(images[0], cmap=plt.cm.gray)
```

# Step 2: Canny Edge Detection

```python
from skimage import filter
edges = [filter.canny(img, sigma=3.5) for img in images] #sigma should be between like 3 and 5 for bold contours
plt.imshow(edges[3], cmap=plt.cm.gray)
```

# Step 3: Corner Detection

```python
def show_images(images, n=4):
    """
        displays n images in a row
    """
    plt.rcParams['figure.figsize'] = 50, 100
    fig, axes = plt.subplots(nrows=1, ncols=n)
    for i, img in enumerate(images[:n]):
        axes[i].imshow(img, cmap=plt.cm.gray)
        axes[i].set_xticks([])
        axes[i].set_yticks([])

show_images(images)
```
