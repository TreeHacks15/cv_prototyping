# Rubiks Cube CV

```python
import sys
sys.path.insert(0, '../')
import numpy 
import scipy
from scipy.misc import imresize
import skimage
import cv2
import matplotlib.pyplot as plt
%pylab inline 

from cv_prototyping import *
```

## Test Data

```python
images = load_rubiks_video()
```

# Resize and Crop
Guaranteed the cube os always in the center of the image

```python
crop = (100, 200)
cropped = [img[crop[0]:-crop[0], crop[1]:-crop[1]] for img in images]
show_images(cropped)
```

## To Grayscale

```python
from cv_prototyping import Preprocess_T

grays = []
for image in cropped:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    grays.append(gray)

show_images(grays)
```

# Corner Detection
Find and mark corners

```python
corner_images = []
corners = []
for ix, gray in enumerate(grays):
    
    out = cv2.cornerHarris(gray,2,3,0.04)

    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
#     out = cv2.cornerSubPix(gray, out, (2,2), (-1,-1), criteria)
    out = cv2.dilate(out,None)
    out = cv2.dilate(out, None)
    out[out>0.01*out.max()] = 255
    corners.append(out)
    
    image = gray.copy()
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    image[out == 255]=[255,0,0]
    corner_images.append(image)

show_images(corner_images)
```

```python
show_images(corners)
```

## Hough Transform from Corners
perform a hough transform on the corner-detected images

```python
line_images = []
for ix, img in enumerate(corners):

#     lines = cv2.HoughLines(image.astype(np.uint8),1,np.pi/180,2)
    lines = cv2.HoughLines(img.astype(np.uint8), 2, np.pi/90, 100)    
#     print lines
    
    line_img = cropped[ix].copy()
    if not lines is None:
        for rho,theta in lines[0]:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))
            cv2.line(line_img,(x1,y1),(x2,y2),(255,0,0),2)

        line_images.append(line_img)

show_images(line_images)
```

# Canny Edge Detection

```python
cannys = []
for gray in grays:
    img = cv2.bilateralFilter(gray, 11, 17, 17)
    canny = cv2.Canny(gray, 35, 150)
    cannys.append(canny)

cannys = [cv2.dilate(canny, np.ones((4,4))) for canny in cannys]
show_images(cannys)
```

## Contours

```python
contours = []
cont_images = []
for ix, canny in enumerate(cannys):
    contours, hierarchy = cv2.findContours(canny,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    img = cropped[ix].copy()
    cv2.drawContours(img, contours, -1, (255,0,0), 3)
    cont_images.append(img)
show_images(cont_images)
```
