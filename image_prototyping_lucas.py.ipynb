# Step 2: Find Plane Edges

```python
from corner_detection import CornerDetector
detector = CornerDetector()
planes = [detector.transform(img) for img in images]
```

```python
fig, axes = show_images(images)

for i, ax in enumerate(axes):
    plane = planes[i]
    xs = [p[0] for p in plane] + [plane[0][0]]
    ys = [p[1] for p in plane] + [plane[0][1]]
    ax.plot(xs,ys)
```
